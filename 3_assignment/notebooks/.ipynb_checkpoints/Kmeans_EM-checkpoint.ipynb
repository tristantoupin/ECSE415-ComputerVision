```python
import numpy as np
from skimage import segmentation, color
from matplotlib import pyplot as plt
import cv2
import math
```

```python
path_img = "home.jpg"
img = cv2.cvtColor(cv2.imread(path_img), cv2.COLOR_BGR2RGB)
```

```python
def plot_images(list_of_images):
    n = len(list_of_images)
    if n == 1:
        plt.imshow(list_of_images[0]); plt.axis('off'); plt.show()
    else:
        r, c = 1, n
        if n > 3:
            c = 3
            r = int(math.ceil(n/3))
    
        fig = plt.figure(figsize=(17, 17))
        for i, (img,name) in enumerate(list_of_images):
            ax = fig.add_subplot(r, c, (i+1))
            ax.set_title(str(name))
            ax.axis('off')
            ax.imshow(img)
```

```python
plot_images([img])
```

# Part A - K Means

```python
def cluster(img, centers, k):
    # get new clusters
    new_cluster = np.empty(shape=(k, img.shape[-1]))
    for i in range(k):
        np.mean(img[centers == i], axis=0, out=new_cluster[i])
    return new_cluster
    
    
def k_means(img, k, iterations = 10, seed = 1234):
    np.random.seed(seed)
    h, w, c = img.shape
    centers = np.random.random_sample((k, c)) * 255.0

    all_imgs = []
    for i in range(iterations):
        # Get distances between each point and each centroid
        dists = np.array([img - c for c in centers])
        sqdists = np.linalg.norm(dists, axis=-1)**2

        # Index of the closest centroid to each data point.
        clusters = np.argmin(sqdists, axis=0)
        all_imgs.append((clusters, "iteration #" + str(i+1)))
        new_centroids = cluster(img, clusters, k)
        centers = new_centroids
    
    plot_images(all_imgs)
```

```python
k_means(img, 2)
```

```python
k_means(img, 3)
```

## Part B - Expectation Maximization

```python
def expected_maximization(img, k, iterations = 10, seed = 1234):
    np.random.seed(seed)
    all_mu = np.random.random((k, img.shape[-1])) * 255
    sigma = np.full((k, img.shape[-1]), (255 / k))
    proba = np.full(k, 1.0 / k)
    
    all_imgs = []
    h, w, c = img.shape
    for m in range(iterations):
        print("On iteration #%s" % (m + 1))
        print("mus: \n%s" % (all_mu))
        print("sigma: \n%s" % (sigma))
        print("probabilities: \n%s\n" % (proba))
        
        # Expextations
        exps = np.zeros(shape=(h, w, k))
        for i in range(h):
            for j in range(w): 
                a = math.e**(-0.5 * ((img[i][j] - all_mu) / sigma)**2)
                b = math.sqrt(2 * math.pi) * sigma
                exps[i][j] = np.linalg.norm(a / b, axis=-1)
                
        exps /= np.sum(exps * proba, axis=-1).reshape((h, w, 1))
        
        # clusterings
        clusters = np.argmax(exps, axis=-1)
        all_imgs.append((clusters, "Iteration #" + str(m+1)))
        
        # Maximization 
        for p in range(k): 
            a = 0
            for i in range(h):
                for j in range(w): 
                    a += exps[i, j, p] * img[i, j]
            all_mu[p] = a / np.sum(exps[:,:,p])
            curr = img[clusters == p]
            sigma[p] = np.sqrt(np.var(curr, axis=0))
            proba[p] = np.size(curr) / np.size(img)
            
    plot_images(all_imgs)
```

```python
expected_maximization(img, 2, 10)
```

```python
expected_maximization(img, 3, 10)
```

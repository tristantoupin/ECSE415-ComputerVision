```python
import numpy as np
import cv2
import sklearn
import os
import math
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
```

```python
path_to_data = "data/"
path_to_people = ["daniel_radcliffe/", "emma_watson/", "rupert_grint/"]
group = path_to_data + "group.png"
```

```python
def plot_images(list_of_images, max_col = 4):
    n = len(list_of_images)
    if n == 1:
        plt.imshow(list_of_images[0], cmap="gray"); plt.axis('off'); plt.show()
    else:
        # get number of columns and rows required
        r, c = 1, n
        if n > max_col:
            c = max_col
            r = int(math.ceil(n/max_col))
    
        fig = plt.figure(figsize=(20, max_col * r))
        for i, (img,name) in enumerate(list_of_images):
            ax = fig.add_subplot(r, c, (i+1))
            ax.set_title(str(name))
            ax.axis('off')
            ax.imshow(img, cmap="gray")
```

```python
group_img = cv2.imread(group, 0)
plot_images([group_img])
```

```python
# Detect all faces in group picture
face_cascade = cv2.CascadeClassifier(os.path.join(cv2.data.haarcascades, 'haarcascade_frontalface_default.xml'))
group_img_box = face_cascade.detectMultiScale(group_img, 1.1, 10)

# Draw a box around each detected face
group_copy = group_img.copy()
for (x, y, w, h) in group_img_box:
    cv2.rectangle(group_copy, (x, y), (x+w, y+h), (255, 255, 255), 2)
```

```python
plot_images([group_copy])
```

```python
def get_train_data(path, title = "Subject"):
    imgs = []
    for file in sorted(os.listdir(path)):
        if "_000" in file:
            imgs.append((cv2.imread(path + file, 0), title + " - " + file[-6:-4]))
        
    return imgs
```

```python
data_0 = get_train_data(path_to_data + path_to_people[0], "Daniel")
data_1 = get_train_data(path_to_data + path_to_people[1], "Emma")
data_2 = get_train_data(path_to_data + path_to_people[2], "Rupert")
```

```python
def get_faces(data, size, s = 1.1):
    faces = []
    for img, title in tqdm(data):
        f = face_cascade.detectMultiScale(img, s, 15)
        if len(f):
            x, y, w, h = f[0]
            img = img[y:y+h, x:x+w]
        img = cv2.resize(img, size)
        faces.append((img, title))
        
    return faces
```

```python
faces_0 = get_faces(data_0, (128, 128), s = 1.05)
```

```python
print(faces_0)
```

```python
plot_images(faces_0)
```
